```python
import os
import numpy as np
```

In this notebook we will calculate some statistics of multiple variance measurements.

```python
dir = "./Zero W"
```

```python
# List files in directory
files = os.listdir(dir)
data = [np.loadtxt(f"{dir}/{i}", skiprows=1) for i in files if i.endswith('.csv')]
data
```

### Calculate mean and standard deviation

```python
for i, j in zip(data, files):
    print(f"Name: {j}, Mean: {np.mean(i)}, Std: {np.std(i)}")
```
